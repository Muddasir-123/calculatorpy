```python
from gtts import gTTS
from IPython.display import Audio ,display
def madlibs():
  name=input("What is your Name ? ")
  city=input("City Name ? ")
  verb=input("Enter your Verb ? ")
  animal=input("Enter a animal name ? ")
  adjective=input("Enter adjective ? ")
  sen=f"{name} goes to {city} where he {verb} a {animal} which was {adjective}."
  tts=gTTS(sen,lang="en")
  tts.save("msg.mp3")
  display(Audio("msg.mp3",autoplay=True))
madlibs()
```

```python
from gtts import gTTS
from IPython.display import Audio,display
age=int(input("What is you age ? "))
if age>=18 and age <=60:
  msg=input("What do you want to listen ? ")
  tts=gTTS(msg,lang="en")
  tts.save("msg.mp3")
  display(Audio('msg.mp3',autoplay=True))
else:
  tts1=gTTS("You are not eligible for see the content",lang="en")
  tts1.save("msg1.mp3")
  display(Audio("msg1.mp3",autoplay=True))
```

```python
from gtts import gTTS
from IPython.display import Audio,display
def calculator(a,b,ope):
  if ope == "+":
    x=f"a + b = {a+b}"
    tts=gTTS(x,lang='en')
    tts.save('msg.mp3')
    display(Audio("msg.mp3",autoplay=True))
  elif ope == "-":
    x = f"A minus B = {a-b}"
    tts1=gTTS(x,lang="en")
    tts1.save("msg1.mp3")
    display(Audio("msg1.mp3",autoplay=True))
  elif ope == "*":
    x = f"a * b = {a*b}"
    tts2=gTTS(x,lang="en")
    tts2.save("msg2.mp3")
    display(Audio("msg2.mp3",autoplay=True))
  elif ope == '/':
    x = f"a / b = {a/b}"
    tts3=gTTS(x,lang="en")
    tts3.save("msg3.mp3")
    display(Audio("msg3.mp3",autoplay=True))
  else:
    tts4=gTTS("Operator is required ! ",lang="en")
    tts4.save("msg4.mp3")
    display(Audio("msg4.mp3",autoplay=True))
a=int(input("Enter your first number ? "))
b=int(input("Enter your second number ? "))
ope=input("Enter your operators ? ")
calculator(a,b,ope)
```
